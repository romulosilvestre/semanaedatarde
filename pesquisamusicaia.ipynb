```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados_grafico = df_geral.sum()
```

```python
dados_grafico.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Quantidade de Ocorrências por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)  # Rotaciona os nomes das categorias para melhor leitura
plt.show()
```

# dados_grafico.plot(kind='bar', figsize=(10, 6), color='skyblue')

- dados_grafico.plot: Esse método é uma função do pandas que permite criar gráficos diretamente de uma Series ou DataFrame.
- kind='bar': O parâmetro kind especifica o tipo de gráfico que você deseja criar. Neste caso, 'bar' indica que queremos um gráfico de barras.
- figsize=(10, 6): O parâmetro figsize define o tamanho da figura do gráfico em polegadas, onde (10, 6) significa 10 polegadas de largura e 6 polegadas de altura.
- color='skyblue': O parâmetro color define a cor das barras no gráfico. Aqui, todas as barras serão coloridas com um tom de azul claro (skyblue).
- plt.title('Quantidade de Ocorrências por Estilo Musical')
- plt.title: Esse método define o título do gráfico. O texto 'Quantidade de Ocorrências por Estilo Musical' será exibido no topo do gráfico, explicando o que ele representa.
- plt.xlabel('Estilos Musicais')
- plt.xlabel: Esse método define o rótulo do eixo X (horizontal). Aqui, o rótulo é 'Estilos Musicais', o que indica que cada barra no gráfico representa um estilo musical diferente.
- plt.ylabel('Quantidade')
- plt.ylabel: Esse método define o rótulo do eixo Y (vertical). - O rótulo 'Quantidade' indica que a altura de cada barra representa a quantidade de ocorrências daquele estilo musical.
- plt.xticks(rotation=45)
- plt.xticks(rotation=45): Esse método ajusta a rotação dos rótulos do eixo X. 
- A rotação de 45 graus é aplicada para facilitar a leitura dos nomes dos estilos musicais, especialmente se eles forem longos.
- plt.show()

    plt.show(): Esse comando exibe o gráfico na tela. Ele é necessário para que o gráfico seja renderizado e mostrado após todas as customizações terem sido aplicadas.
